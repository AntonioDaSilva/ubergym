```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open("../data/average_passengers.pkl", "rb") as f:
    average_passengers = pickle.load(f)

with open("../data/passenger_generation_probabilities.pkl", "rb") as f:
    passenger_generation_probabilities = pickle.load(f)
```

```python
plt.figure(figsize=(15,8))
plt.title("Average passengers per node")
plt.xlabel("Step")
plt.ylabel("Passengers")

start_step = 500


for i in range(len(average_passengers)):
    pi = average_passengers[i,:]
    pi = pi[start_step:]
    plt.plot(np.arange(len(pi)), pi, label = str(i))
    plt.plot(np.arange(len(pi)), [passenger_generation_probabilities[i]]*len(pi), label = str(i) + ' (real) ')

plt.legend()
```

```python
total_average_passengers = average_passengers.sum(axis=0)
total_average_passengers.shape
```

```python
plt.figure(figsize=(15,8))
plt.title("Average passengers")
plt.xlabel("Step")
plt.ylabel("Passengers")

plt.plot(np.arange(len(total_average_passengers)), total_average_passengers, label = 'total')
plt.plot(np.arange(len(total_average_passengers)), [passenger_generation_probabilities.sum()]*len(total_average_passengers), label = 'total (real)')
plt.legend()
```
