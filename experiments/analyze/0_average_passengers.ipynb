```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open("../data/average_passengers.pkl", "rb") as f:
    average_passengers = pickle.load(f)


plt.title("Average passengers per node")
plt.xlabel("Step")
plt.ylabel("Passengers")


for i in range(len(average_passengers)):
    pi = average_passengers[i,:]
    plt.plot(np.arange(len(pi)), pi, label = str(i))

plt.legend()
plt.show()
```
